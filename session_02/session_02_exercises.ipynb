### B5 · Compounding, one day at a time  ★★★☆☆

Returns do not add, they **compound**: each day multiplies what you had by $(1+r)$. Start a
`value` at **1.0** and multiply it by `(1 + r)` once per return, then subtract 1 to get the total
return over the week.

$$r_\text{total} = \prod_i (1+r_i) - 1$$

````python
returns = [0.012, -0.004, 0.008, -0.011, 0.006]

value = ...
for r in returns:
    ...
compounded = ...
print(compounded)
````

<details>
<summary>💡 Hint 1</summary>

Start at `value = 1.0` (the neutral value for multiplying, the way 0 is for adding).

</details>

<details>
<summary>💡 Hint 2</summary>

Inside: `value *= (1 + r)`. After the loop: `compounded = value - 1`.

</details>

<details>
<summary>✅ Solution</summary>

```python
returns = [0.012, -0.004, 0.008, -0.011, 0.006]

for r in returns:
```

`0.010868480889344001`, that is **1.0868%**, against the naive sum of 1.1000%. The accumulator starts at 1 and uses `*=` because the neutral value for multiplication is one, not zero.

</details>

## 🔍 L · Functions nobody taught you

*No course covers everything, and this one has deliberately covered very little. The skill that outlasts any syllabus is being able to find a tool you have never seen, work out what it does, and use it correctly.*

**Three places to look**

1. **`help(name)`**, Python's own manual, built into the language. It works offline and it is always the version you are actually running. Every exercise below starts here.
2. **The official documentation**, at `docs.python.org`. Longer, with examples and the reasoning behind a function.
3. **Ask an AI to explain a function** (what it does, what its arguments mean). Asking for an explanation you then verify is good practice. Asking it to write your answer is how you fail the exam.

**Reading a signature.** `help` opens with a line like `round(number, ndigits=None)`. That tells you the order of the arguments and that anything shown with an `=` has a **default**, so you may leave it out. It is the same rule you met when you wrote `def grow(price, rate=0.05)` in H4.

Ignore all of it. Read the **signature** at the top and the **first paragraph** underneath, and stop there. Learning what to skip is half of reading documentation.

**Drills:** `help`, `sorted`, `abs`, `.keys()`, `.values()`, `.items()`, `enumerate`, `zip`.

### M4 · The comparison that fails  ★★★★☆  · revisits S1

⚠️ **Run the cell below.** Three returns of 0.1 are added up in a loop, and the total is compared
with 0.3. Python says they are **not** equal.

Nothing is broken. Print `total` to see why, then repair the comparison so it says `True`.

````python
returns = [0.1, 0.1, 0.1]

total = 0
for r in returns:
    total = total + r

print(total)
print(total == 0.3)
````

<details>
<summary>💡 Hint 1</summary>

Look at what `total` actually prints. Decimals that are simple in base 10 are not simple in base 2.

</details>

<details>
<summary>💡 Hint 2</summary>

Compare rounded values instead: `round(total, 10) == round(0.3, 10)`.

</details>

<details>
<summary>✅ Solution</summary>

```python
returns = [0.1, 0.1, 0.1]

total = 0
for r in returns:
    total = total + r

print(total)
print(round(total, 10) == round(0.3, 10))
```

`0.30000000000000004`, so the raw comparison is `False`. This is not a Python defect, it is how binary floating point works in every language.

**Never test two computed decimals with `==`.** Round both sides first, or ask whether the difference is small. Every language you will ever use has this trap, and this is the cheapest possible place to meet it.

</details>
