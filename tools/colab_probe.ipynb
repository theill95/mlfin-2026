# Round 2: can any HTML layout carry real mathematics?

Round 1 settled that the table **type** does not matter. Colab refuses to typeset maths inside table markup, markdown or HTML alike, and it does not know the `\(...\)` delimiters at all.

**Read N first.** It is the diagnostic, and it decides the rest:

- if the fraction in **N** renders as real mathematics, a two-column layout is possible and **G** or **H** will give you the table look back
- if **N** shows raw `\dfrac` text, then no HTML layout can carry LaTeX in Colab, and the honest choice is between **E** and **F** from round 1


---

## N. The diagnostic: maths inside a plain `<div>` and `<span>`

No table anywhere. If Colab typesets maths in raw HTML at all, it happens here.

<div>Inside a div: $\dfrac{a}{b}$ and $\sum_i x_i^2$</div>
<p>Inside a p: $R^2=1-\dfrac{SS_{res}}{SS_{tot}}$</p>
<span>Inside a span: $\sqrt{\text{MSE}}$</span>

## G. Two columns with CSS grid, no table markup

This is the one that would give you back what you liked: aligned columns, real mathematics, and not a `<table>` in sight.

<div style="display:grid;grid-template-columns:max-content 1fr;gap:0.45rem 1.5rem;align-items:baseline">

</div>
