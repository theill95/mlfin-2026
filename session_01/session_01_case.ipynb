### Q7 — Make the risk call

Now the analyst's judgment. A rough-and-ready measure of how *bumpy* a

For each stock: `(max(closes) - min(closes)) / closes[0]`.

</details>


<details>
<summary>Hint 2</summary>

</details>


<details>
<summary>Solution</summary>

```python

ko_rel_range = (max(ko_closes) - min(ko_closes)) / ko_closes[0]

print(f"Coca-Cola relative range: {ko_rel_range:.2%}")

# Apple looked far riskier: its relative range (~51%) dwarfs
# Coca-Cola's (~26%). But 'riskier' is NOT the same as 'lower return' --
# Apple was both bumpier AND higher-returning. Risk and return are two
# different measurements, and this course keeps them carefully apart.
```

**Apple relative range 51.22%, Coca-Cola 26.15%.** Apple swung across a much wider band relative to its price — the bumpier ride — even though it also earned much more. This is the report's headline, and it plants a seed for the whole course: *risk* (how much a price moves) and *return* (how much it earns) are distinct. This crude range measure becomes a proper **volatility** measure in Session 2, and by Session 4 we will ask — carefully — whether movement like this can be **predicted** at all.

</details>
